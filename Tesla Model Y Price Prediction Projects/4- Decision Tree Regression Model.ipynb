# **Data Cleaning Check Template**
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:


```python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
```

# **Load dataset**
Apply Data Cleaning Check Template

```python
dataset = pd.read_csv("/kaggle/input/datasets/aiwithcagri/multiple-linear-regression-tesla-model-y-price/teslay.csv")
data_quality_report(dataset)
```

# **Understaning the outliers** 

### km: 6 outliers

```python
# Calculate fences for Price
Q1 = dataset['km'].quantile(0.25)
Q3 = dataset['km'].quantile(0.75)
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Filter the dataframe to show ONLY the outliers
km_outliers = dataset[(dataset['km'] < lower_fence) | (dataset['km'] > upper_fence)]

# Display the rows
print(km_outliers)
```

# **Cleaning the data** 

### Remove km outliers (6) & remove the duplication (4)

```python
# Remove rows that exist in km_outliers from the original dataset
dataset_clean = dataset.drop(km_outliers.index).copy()

# Step 2: Remove duplicates FROM the cleaned dataset
dataset_clean = dataset_clean.drop_duplicates().copy()

# Final result
print(dataset_clean)
```

# **Select Independent Variable (X) and Dependent Variable (y)**

📌 Note: The concepts of Independent and Dependent variables were discussed earlier.

If you would like to review or gain a clearer understanding, please refer to this section: [Go to Independent variable VS Dependent variable Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
# Separate independent variable(features (X)) by selecting all rows and all columns except the last one
X = dataset_clean.iloc[:, :-1].values
# Separate the dependent variable (Target(y)) by selecting the last column
y = dataset_clean.iloc[:, -1].values
```

# **Encoding categorical data**

## Encoding the Indpendent Variable Using One Hot Encoding

📌 Note: I explained One-Hot Encoding earlier.

If you missed it or want a deeper understanding of how it works, you can find it here: [Go to One-Hot Encoding Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'color' column (index 2) while keeping other features
# Initialize ColumnTransformer to handle categorical data
# [2] targets the 'color' column (index 2)
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse_output=False),[2])], remainder = 'passthrough')
# Execute transformation and convert to NumPy ndarray for model training
X = np.array(ct.fit_transform(X))
```

# **Splitting the dataset into the Train set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train , y_test = train_test_split(X, y , test_size = 0.2, random_state = 0)
```

```python
print (X_test)
```

```python
print(f"Rows: {X_test.shape[0]}, Columns: {X_test.shape[1]}")
```

# **Decision Tree Regression Model**

## Training the Building Decision Tree Regression model on the Training set

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=3, min_samples_split=5, min_samples_leaf=3, random_state = 0)
regressor.fit(X_train, y_train)
```

1- `max_depth`: The maximum number of levels in the tree.

- Depth = how many splits from root → leaf
- A deeper tree = more complex model

**Effect:**

- Small depth → underfitting (too simple)
- Large depth → overfitting (too complex)

---

2- `min_samples_split`: The minimum number of samples required to split a node.

- The model will NOT split a node unless it has at least this many samples.

**Effect:**

- Small value → more splits → complex tree
- Large value → fewer splits → simpler tree

---

3- `min_samples_leaf`: The minimum number of samples allowed in a leaf (final node)

- After splitting, each leaf must have at least this many samples.

**Effect:**

- Small value → overfitting risk
- Larger value → smoother predictions


# **Preticting the Test set results**

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# **Evaluating the Model Performance**

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# **Important Insight**

1. The initial Decision Tree model produced a negative R² score (-0.35), indicating a complete failure to generalize beyond training data.

2. After applying regularization constraints (max_depth=3, min_samples_split=5, min_samples_leaf=3), performance improved significantly to R² = 0.49.

3. This confirms that the issue was not the model itself, but uncontrolled model complexity leading to overfitting.

4. The improvement demonstrates that capacity control is critical, especially when working with small datasets (245 samples).
5. In comparison, SVR achieved the highest performance (R² = 0.53) when preprocessing was applied selectively, reinforcing that both model choice and preprocessing strategy directly impact performance.

---

# **Conclusion**

The experiments show that model performance is highly sensitive to both preprocessing strategy and model complexity.

The Decision Tree model initially failed due to overfitting, as indicated by a negative R² score (-0.35). By introducing constraints to control model complexity, performance improved significantly to 0.49, confirming that generalization depends on proper regularization rather than model selection alone.

Similarly, SVR performance varied substantially depending on preprocessing choices. Scaling all features led to poor results (0.23), while selectively scaling only high-magnitude numerical features improved performance to 0.53, the best result among all models.

These results highlight that effective machine learning is driven by informed design decisions, including selective preprocessing and controlled model capacity, rather than applying standard techniques without validation.
